# Customer Personality Analysis - Data Cleaning

## Objective
The goal of this project is to clean the Customer Personality Analysis dataset and transform it into a reliable, analysis-ready dataset.

The cleaning process includes:
- Data quality assessment
- Missing value treatment
- Duplicate removal
- Data standardization
- Outlier detection
- Data type correction
- Before vs After comparison
- Exporting the cleaned dataset

```python
import pandas as pd
```

```python
before_summary = {
    "Rows": df.shape[0],
    "Columns": df.shape[1],
    "Missing Values": df.isnull().sum().sum(),
    "Duplicate Rows": df.duplicated().sum(),
    "Incorrect Data Types": 3
}
print(before_summary)
```

Before cleaning the dataset, there existed 2240 rows, 29 columns, 24 missing values no duplicate rows, 3 columns with incorrect data type

```python
#Converting Columns to correct format
df['ID'] = df['ID'].astype(str)
df['Year_Birth'] = pd.to_datetime(df['Year_Birth'],dayfirst=True).dt.year
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'],dayfirst=True)
print(df.info())
```

```python
#Filling the null records
median_income = df['Income'].median()
df['Income'] = df['Income'].fillna(median_income)
```

Income contained missing values. Since income distributions are often right-skewed and sensitive to extreme values, the median was used for imputation instead of the mean.

```python
#Checking for unique values
print(df['Marital_Status'].unique())
```

We see that the Marital status column consists of various unique values most of which are unnecessary and can be accomodated in already existing names which is what we'll try to achieve in the next cell

```python
df['Marital_Status'] = df['Marital_Status'].replace({
    'Absurd':'Other',
    'YOLO': 'Other',
    'Together': 'Married',
    'Alone': 'Single',
    'Widow': 'Single'
})
print(df['Marital_Status'].unique())
```

The values of the Marital Status column have been trimmed down to: Single, Married, Divorced, Other

```python
#Outlier Detection
numeric_cols = df.select_dtypes(include=np.number).columns

#Loop through them using the IQR method.
for col in numeric_cols:

    Q1 = df[col].quantile(.25)
    Q3 = df[col].quantile(.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR

    outliers = ((df[col] < lower) | (df[col] > upper)).sum()

    print(col, outliers)
```

### Outlier Detection

Outliers were identified using the Interquartile Range (IQR) method. A few columns, especially those related to customer income and product spending, contained values outside the expected range. After reviewing these values, I decided to keep them because they appear to reflect genuine customer behaviour rather than data entry mistakes. For example, customers with very high incomes or spending habits are realistic and could provide valuable insights during analysis.

Binary columns such as `AcceptedCmp1`, `Response`, and `Complain` were not treated as outliers because they only contain two possible values (0 and 1), making the IQR method unsuitable for those variables.


```python
after_summary = {
    "Rows": df.shape[0],
    "Columns": df.shape[1],
    "Missing Values": df.isnull().sum().sum(),
    "Duplicate Rows": df.duplicated().sum(),
    "Incorrect Data Types": 0
}
print(after_summary)
```

After cleaning the dataset, there are 2240 rows, 29 columns,no missing values, no duplicate rows, 0 columns with incorrect data type.

```python
summary = pd.DataFrame({
    "Before Cleaning": before_summary,
    "After Cleaning": after_summary
})

print(summary)
```
